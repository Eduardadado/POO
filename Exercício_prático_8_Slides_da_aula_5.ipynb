```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('vendas_brasil_aula3_bruto.csv')
df.head()
```

```python
df_uf = df.groupby('uf')['receita'].sum().sort_values(ascending=False)

plt.figure()
df_uf.plot(kind='bar')
plt.title('Receita por UF')
plt.xlabel('UF')
plt.ylabel('Receita')
plt.show()
```

```python
df_canal = df.groupby('canal')['lucro'].sum().sort_values(ascending=False)

plt.figure()
sns.barplot(x=df_canal.index, y=df_canal.values)
plt.title('Lucro por Canal')
plt.xlabel('Canal')
plt.ylabel('Lucro')
plt.show()
```

```python
df_tempo = df.groupby('ano_mes')['receita'].sum()

plt.figure()
df_tempo.plot()
plt.title('Receita ao longo do tempo')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure()
sns.histplot(df['margem_lucro'], bins=20)
plt.title('Distribuição da Margem de Lucro')
plt.show()
```
